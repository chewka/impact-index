```python
from bs4 import BeautifulSoup
import requests
from lxml import html
from urllib.request import urlopen

```

```python
url = "https://www.monarchsf.com/"
session = requests.Session()
resp = session.get(url)
html = resp.text

soup = BeautifulSoup(html, 'lxml')
title = soup.find('h1', class_ = 'event-name headliners')
date = soup.find('span', class_ = 'dates')
datos = soup.findAll('span', class_ = 'dates')
```

```python
events = []
for h in soup.find_all('h1', class_ = 'event-name headliners'):
    a = h.text
    events.append(a)
```

```python
events
```

```python
dates = []
for d in soup.findAll('span', class_ = 'dates'):
    date = d.text
    dates.append(date)
```

```python
event_dict = {}
for i in range(len(events)):
    event_dict[events[i]] = dates[i]
    
print(event_dict)
```

```python
url = 'http://1015.com/calendar/'
session = requests.Session()
resp = session.get(url)
html = resp.text

html.replace("\r", " ").replace("\n", " ")

soup2 = BeautifulSoup(html, 'lxml')

for e in soup2.findAll('br'):
    e.extract()

events = []
for h in soup2.find_all('h3', class_ = 'truncate'):
    a = h.text.rstrip()
    events.append(a)
    
dates = []
for d in soup2.findAll('h1'):
    date = d.text
    dates.append(date)
    
```

```python
soup2
```

```python
#events = soup2.find_all('span', class_ = 'truncate')
```

```python
for i in range(len(events)):
     events[i] = events[i].replace('\r\n', ' ')

events
```

```python
dates = dates[1:]
```

```python
for i in range(len(dates)):
     dates[i] = dates[i].rstrip().replace('\n\n', '')
```
