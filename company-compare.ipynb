```python
# Python code to generate 
# random numbers and 
# append them to a list 
import random 
  
# Function to generate 
# and append them  
# start = starting range, 
# end = ending range 
# num = number of  
# elements needs to be appended 
def Rand(start, end, num): 
    res = [] 
  
    for j in range(num): 
        res.append(random.randint(start, end)) 
  
    return res 
  
# Driver Code 
num = 30
start = 0
end = companies.shape[0]
randoms = (Rand(start, end, num)) 
print(randoms)
```

```python
def find_randoms(companies, zero):

    # Driver Code 
    num = 6
    start = 0
    end = companies.shape[0]

    def Rand(start, end, num): 
        res = [] 

        for j in range(num): 
            res.append(random.randint(start, end)) 

        return res 

    randoms = Rand(start, end, num)
    for i in randoms:
        zero = zero + 1
        print(str(zero) + ' ' + str(companies.at[i, 'company_name']))
        print('\nCB')
        # + str(companies.at[i, 'jobs_created'])#, 'funding','jobs_created', 'age_years', 'founding_date'])
        print('founders: ' + str(companies.at[i, 'founders']))
        print('funding: ' + str(companies.at[i, 'funding']))
        print('IPO: ' + str(companies.at[i, 'ipo']))   
        print('jobs created: ' + str(companies.at[i, 'jobs_created']))
        
        print('city: ' + str(companies.at[i, 'city']))
        print('\nPB \nfounders: \nfunding: \nemployees: \n \nMM \nfounders: \nfunding: \nemployees: \n \n')
        
    

        print('------- \n')

```
