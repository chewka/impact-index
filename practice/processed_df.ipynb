```python
import pandas as pd
from pathlib import Path

PROJ_ROOT = Path().resolve().parent
PRC_DIR = PROJ_ROOT / 'data' / 'processed'

company_df = pd.read_csv(PRC_DIR / 'processed_company_df.csv', index_col=0)
```

```python
company_df
```

```python
industries = []
for each in company_df['category']:
    cats = each.replace('[', '').replace(']', '')
    cats = cats.split(', ')
    cats = [x.replace("'", '') for x in cats]
    #print(cats)
    for c in cats:
        if c not in industries:
            if c != '':
                industries.append(c)
```
