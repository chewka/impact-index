```python
import pandas as pd
from pathlib import Path

PROJ_ROOT = Path().resolve().parent
PRC_DIR = PROJ_ROOT / 'data' / 'processed'

state_df = pd.read_csv(PRC_DIR / 'metro_df.csv', index_col=0)
```

```python
state_df.head()
```

```python
cities = []
for each in state_df['comb_metro']:
    if each not in cities:
        cities.append(each)
```
